```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
# Reading in the csv file created from the WeatherPy portion.
data_raw = pd.read_csv("..\WeatherPy\output_data\cities.csv")
data_raw = data_raw.set_index("city_id")
data_raw.index.name = None
data_raw.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# An API key assignment is necessary to configure gmaps before use.  
gmaps.configure(api_key=g_key)

# Combining latitude and longitude to get coordinates.
locations = data_raw[["lat", "lng"]]
# The heatmap is going to be weighted on humidity.
weights = data_raw["humidity"]

# Zoom of 1.75 is to get the full image of the world, but not too much more.
# Center 10,0 helped cut off some unneeded ocean near the south pole.
fig = gmaps.figure(center=(10, 0), zoom_level=1.75)

# Adding heatmap_layer with our set locations and weights.  A higher max_intensity is used so that the
# entire world doesn't just show red everywhere.
fig.add_layer(gmaps.heatmap_layer(locations, weights=weights, max_intensity=125, dissipating=True))
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

* For example:
  * A max temperature lower than 80 degrees but higher than 70.

  * Wind speed less than 10 mph.

  * Zero cloudiness.

  * Drop any rows that don't contain all three conditions. You want to be sure the weather is ideal.
  
  **Note:** Feel free to adjust to your specifications but be sure to limit the number of rows returned by your API requests to a reasonable number.

```python
# Set criteria for preferred weather.
humidity_limit = 80
wind_speed_limit = 5
cloudiness_limit = 10
max_temp_min = 70
max_temp_max = 95


# Using .loc to filter the dataframe down with chosen criteria.  These changes are saved to a new dataframe.
w_criteria = data_raw.loc[data_raw["humidity"] < humidity_limit, :]
w_criteria = w_criteria.loc[w_criteria["wind_speed"] < wind_speed_limit, :]
w_criteria = w_criteria.loc[w_criteria["cloudiness"] < cloudiness_limit, :]
w_criteria = w_criteria.loc[(w_criteria["max_temp"] > max_temp_min) & (data_raw["max_temp"] < max_temp_max), :]
w_criteria
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# The criteria dataframe is saved to hotel_df as requested.  
hotel_df = w_criteria
# Adding blank column "Hotel Name".
hotel_df["Hotel Name"] = ""
hotel_df
```

```python
# Base url of the google places api.
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?"

# Creating dictionary of the parameters desired for our search.  
# Location is added later with a new value each time it's looped through.
param = {

    "type": "lodging",
    "radius": 5000,
    "key": g_key
    
}

# looping through each row of the hotel_df dataframe.
for index, row in hotel_df.iterrows():
    
    # Assigning variables to a column of the current row.
    lat = row["lat"]
    lng = row["lng"]
    city = row["city"]
    # Concatinating latitude and longitude into a string and then assigning it to the
    # location list in our param dictionary.
    param["location"] = f"{lat},{lng}"
    

        
    # assemble url and make API request
    print(f"Retrieving Results for Index {index}: {city}")
    response = requests.get(base_url, params=param).json()
    
    # extract results
    results = response["results"]
    
    # Attempt to do the following.
    try:
        print(f"Closest Hotel in {city} is {results[0]['name']}.")
        
        # In the current row of the loop, add value of results[0]['name'] to the ["Hotel Name"] column in hotel_df.
        row["Hotel Name"] = results[0]['name']
        
    
    # If one of these two errors come up do the following.
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
    print("------------")
```

```python
# Validating Hotel Names were printed to dataframe.
hotel_df.head()
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{city}</dd>
<dt>Country</dt><dd>{country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["lat", "lng"]]
```

```python
# Add marker layer ontop of heat map
fig.add_layer(gmaps.marker_layer(locations, info_box_content=hotel_info))


# Display figure
fig
```
